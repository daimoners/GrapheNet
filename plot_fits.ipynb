# Production plots

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data_df = pd.read_csv('/home/tommaso/git_workspace/GrapheNet/data_GO/training_dataset_reference/dataset.csv', header=0)
data_G_df = pd.read_csv('/home/tommaso/git_workspace/GrapheNet/data_G/training_dataset_reference/dataset.csv', header=0)
```

## Electron affinity GO

```python
bound = [-7.5, -4.2]

target = "electron_affinity"

df = pd.read_csv(f"/home/tommaso/git_workspace/GrapheNet/data_GO/training_dataset_reference/models/inceptionresnet/last_high_reduced_layers_lr_workstation_optimized/{target}_new/{target}_prediction_results.csv", header=0)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 5), gridspec_kw={'width_ratios': [4, 1]})

# Plot the scatter plot on the left subplot
ax1.scatter(df['electron_affinity_predicted'], df['electron_affinity_real'], s=20, alpha=0.3,linewidth=0)
ax1.set_xlabel('EA (eV) - predicted',size=20)
ax1.set_ylabel('EA (eV) - true',size=20)

ax1.tick_params(axis='both', which='major', labelsize=20)

#ax1.axis('equal')
ax1.set_xlim(bound)
ax1.set_ylim(bound)
ax1.set_box_aspect(1)

# Add y=x reference axis
lowbound1 = bound[0]*0.98
lowbound2 = bound[1]*1.02
lowbound = [lowbound1, lowbound2]
ax1.plot(lowbound, lowbound)

# Plot the histogram on the right subplot
ax2.set_ylim(bound)
hist_data=data_df['electron_affinity']
w=0.01
ax2.hist(hist_data, orientation='horizontal',bins=np.arange(min(hist_data), max(hist_data) + w, w),linewidth=0,alpha=0.6)
ax2.tick_params(axis='both', which='major', labelsize=20)
ax2.set_xticks([])
ax2.yaxis.tick_right()
# Adjust the position of the subplots
fig.subplots_adjust(wspace=-0.4)

# Show the plot
plt.show()
```

## Formation energy - GO

```python
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)

bound = [-15, -1]

target = "formation_energy"

df = pd.read_csv(f"/home/tommaso/git_workspace/GrapheNet/data_GO/training_dataset_reference/models/inceptionresnet/last_high_reduced_layers_lr_workstation_optimized/{target}_new/{target}_prediction_results.csv", header=0)

# Create a figure with two subplots
fig, (axe1, axe2) = plt.subplots(ncols=2, figsize=(10, 5), gridspec_kw={'width_ratios': [4, 1]})

# Plot the scatter plot on the left subplot
axe1.scatter(df['formation_energy_predicted']/1000, df['formation_energy_real']/1000, s=20, alpha=0.1,linewidth=0)
axe1.set_xlabel('Eform (eV*1E3) - predicted',size=20)
axe1.set_ylabel('Eform (eV*1E3) - true',size=20)

axe1.tick_params(axis='both', which='major', labelsize=20)
axe1.xaxis.set_major_locator(MultipleLocator(5.0))
axe1.yaxis.set_major_locator(MultipleLocator(5.0))
axe1.axis('equal')
axe1.set_xlim(bound)
axe1.set_ylim(bound)
axe1.set_box_aspect(1)

# Add y=x reference axis
lowbound1 = bound[0]*0.95
lowbound2 = bound[1]*1.35
lowbound = [lowbound1, lowbound2]
axe1.plot(lowbound, lowbound)

# Plot the histogram on the right subplot
hist_ene_data=data_df['formation_energy']/1000
w=0.02
axe2.hist(hist_ene_data, orientation='horizontal',bins=np.arange(min(hist_ene_data), max(hist_ene_data) + w, w),linewidth=0,alpha=0.6)
axe2.set_ylim(bound)
#axe2.set_xlim([0,20])
axe2.tick_params(axis='both', which='major', labelsize=20)
#axe2.set_xticks([])
axe2.yaxis.tick_right()
# Adjust the position of the subplots
fig.subplots_adjust(wspace=-0.4)

# Show the plot
plt.show()
```
