```python
def r2_score(y_pred: list, y_true: list):
    y_pred = np.array(y_pred)
    y_true = np.array(y_true)
    mean_y = np.mean(y_true)
    SSR = np.sum((y_pred - y_true) ** 2)
    SST = np.sum((y_true - mean_y) ** 2)
    return 1 - SSR / SST
```

## Electron affinity GO

```python
bound = [-7.5, -4.2]

target = "electron_affinity"

df = pd.read_csv(f"/home/tommaso/git_workspace/GrapheNet/data_GO/training_dataset_reference/models/inceptionresnet/last_high_reduced_layers_lr_workstation_optimized/{target}_new/{target}_prediction_results.csv", header=0)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 5), gridspec_kw={'width_ratios': [4, 1]})

# Plot the scatter plot on the left subplot
ax1.scatter(df['electron_affinity_predicted'], df['electron_affinity_real'], s=20, alpha=0.3,linewidth=0)
ax1.set_xlabel('EA (eV) - predicted',size=20)
ax1.set_ylabel('EA (eV) - true',size=20)

ax1.tick_params(axis='both', which='major', labelsize=20)

#ax1.axis('equal')
ax1.set_xlim(bound)
ax1.set_ylim(bound)
ax1.set_box_aspect(1)

# Add y=x reference axis
lowbound1 = bound[0]*0.98
lowbound2 = bound[1]*1.02
lowbound = [lowbound1, lowbound2]
ax1.plot(lowbound, lowbound)

# Plot the histogram on the right subplot
ax2.set_ylim(bound)
hist_data=data_df['electron_affinity']
w=0.01
ax2.hist(hist_data, orientation='horizontal',bins=np.arange(min(hist_data), max(hist_data) + w, w),linewidth=0,alpha=0.6)
ax2.tick_params(axis='both', which='major', labelsize=20)
ax2.set_xticks([])
ax2.yaxis.tick_right()
# Adjust the position of the subplots
fig.subplots_adjust(wspace=-0.4)

# Show the plot
plt.show()

print(f"R2 fit: {r2_score(df['electron_affinity_predicted'].to_list(),df['electron_affinity_real'].to_list())}")
```
